```python
import numpy as np

matrices = [np.random.randint(-10, 11, size=(3, 3)) for _ in range(3)]

for i, A in enumerate(matrices, start=1):
    print("\n" + "="*40)
    print(A)

    try:
        A_inv = np.linalg.inv(A)
        print(A_inv)

        product = A @ A_inv
        print(product)
    except np.linalg.LinAlgError:
        print("this matrix has no inverse")
```

```python
import pandas as pd

df=pd.read_csv("/Users/arpithacoorg/Desktop/spring 2026/machine learning/kc_house_data.csv")
df=df.drop(columns=["id","date","zipcode"])
y=df["price"]
x=df.drop(columns=["price"])
summary = pd.DataFrame({
    "mean": x.mean(),
    "min": x.min(),
    "max": x.max(),
    "variance": x.var(ddof=0)
})
summary
print("Lowest average feature:", summary["mean"].idxmin(),
     summary.loc[summary["mean"].idxmin(), "mean"])
print("Highest average feature:", summary["mean"].idxmax(),
      summary.loc[summary["mean"].idxmax(), "mean"])
print("Lowest variance feature:", summary["variance"].idxmin(),
      summary.loc[summary["variance"].idxmin(), "variance"])
print("Highest variance feature:", summary["variance"].idxmax(),
      summary.loc[summary["variance"].idxmax(), "variance"])
```

```python
corr_table = x.corrwith(y).to_frame(name="correlation")
corr_table = corr_table.sort_values("correlation", ascending=False)
corr_table

positive_features = corr_table[corr_table["correlation"] > 0]
print("Positively correlated features:")
print(list(positive_features.index))
print("Feature with highest positive correlation:",
      corr_table.index[0],
      corr_table.iloc[0, 0])
```
